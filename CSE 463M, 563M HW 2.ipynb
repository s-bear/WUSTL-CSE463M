**1.** Design suitable circuits in Cadence that will allow you to compute the: voltage threshold ($V_T$), $k^\prime$, and substrate bias coefficient ($\gamma$) for an NMOS transistor. The transistor aspect ratios are $W=9\,\mu\text{m}$ and $L=9\,\mu\text{m}$. Print the circuits that you design to extract these 3 parameters. Simulate the circuit and print the simulated results. Indicate on your plots how you computed the 3 parameters.

**Solution:** Use the following circuit, which holds the NMOS in saturation mode by shorting the gate and drain together (so that $V_{GS}=V_{DS}$). The first-order model for the system is thus
$$I_D\left(V_{GS},V_{SB}\right)=\frac{1}{2}k_n^\prime\frac{W}{L}\left(V_{GS}-V_T \left( V_{SB} \right) \right)^2)$$
$$V_T \left(V_{SB}\right)=V_{T0}+\gamma\left(\sqrt{V_{SB}+|2\phi_F|}-\sqrt{|2\phi_F|} \right)$$
Taking the square root yields
$$\sqrt{I_D}=\sqrt{\frac{k_n^\prime W}{2 L}}\left(V_{GS}-V_T\left(V_{SB}\right)\right)=\sqrt{\frac{k_n^\prime W}{2 L}}V_{GS}-\sqrt{\frac{k_n^\prime W}{2 L}}V_T\left(V_{SB}\right)$$ which takes the form of $$y = mx + b$$ where $y = \sqrt{I_D}$, $x = V_{GS}$, $m = \sqrt{k_n^\prime W / (2L)}$, and $b = -\sqrt{k_n^\prime W/(2L)}V_T\left(V_{SB}\right)$. The $x$-intercept will be $x|_{y=0}=-b/m=V_T\left(V_{SB}\right)$.
This informs our procedure for finding the desired quantities. We can use the slope of the curve to find $k_n^\prime=2 m^2 L/W$, and the $x$-intercept when $V_{SB}=0$ to find $V_{T0}$. Finding $\gamma$ requires setting $V_{SB} > 0$. The most convenient value comes from solving 
$$V_T-V_{T0}=\gamma\left(\sqrt{V_{SB}+|2\phi_F|}-\sqrt{|2\phi_F|}\right)=\gamma$$
or
$$\left(\sqrt{V_{SB}+|2\phi_F|}-\sqrt{|2\phi_F|}\right)=1$$
$$V_{SB}=1+2\sqrt{|2\phi_F|}$$
$\phi_F$ can be determined from the model parameters $T=300\,\text{K}$, and $N_A=1.7\times10^{17}$ using $\phi_F=(kT/q)\,\text{ln}(n_i/N_A)=-0.421\,\text{V}$. So $V_{SB}=2.835\,\text{V}$.

```python
#set up environment
%pylab inline
from IPython.display import Image
rc('font',size=20)
```

```python
#the nmos circuit
Image(filename='nmos_ckt.png')
```

```python
#The circuit was simulated as a DC sweep of VGS from 0 V to 5 V, first with VSB = 0 V, then with VSB = 2.835 V
#import data from CSV file (saved from the Results Browser Table View)
vgs,Id_n_min,Id_n_big,Id_n_min_vsb,Id_n_big_vsb = loadtxt('nmos_results.csv',delimiter=',',skiprows=1,unpack=True)
vsg,Id_p_big,Id_p_min,Id_p_big_vsb,Id_p_min_vsb = loadtxt('pmos_results.csv',delimiter=',',skiprows=1,unpack=True)
# *_min (denoting minimum size transistor) data are for problem 4
```

```python
#make a function because we're going to do this several times:
def find_parameters(vgs, Id, Id_v):
    #Find where the transistor starts conducting. We can then do a linear fit beyond that point to get what we need.
    #The transistor does not fit the model exactly, so we only use a short region past where it starts conducting.
    Id_lo = find(Id > 5e-6)[0] #first index where Id > 5 uA
    Id_hi = find(vgs > vgs[Id_lo]+1)[0] # up 1 V, since the model doesn't fit very well at larger voltages
    Id_v_lo = find(Id_v > 5e-6)[0]
    Id_v_hi = find(vgs > vgs[Id_v_lo]+1)[0]

    fit = polyfit(vgs[Id_lo:Id_hi],sqrt(Id[Id_lo:Id_hi]),1) #1st order polynomial fit, coefficients in decreasing order
    fitf = poly1d(fit) #function that evaluates the fit

    fit_v = polyfit(vgs[Id_v_lo:Id_v_hi],sqrt(Id_v[Id_v_lo:Id_v_hi]),1)
    fitf_v = poly1d(fit_v)

    #find parameters:
    Vt0 = -fit[1]/fit[0] # = -b/m, Vsb = 0
    Vt1 = -fit_v[1]/fit_v[0] # = -b/m, Vsb = 2.835
    gamma = Vt1-Vt0
    kn = 2*fit[0]**2 # = 2*m^2*L/W
    
    return (Vt0, Vt1, gamma, kn, fitf, fitf_v)

Vt0, Vt1, gamma, kn, fit, fit_v = find_parameters(vgs, Id_n_big, Id_n_big_vsb)

#print our results
print('NMOS: W = L = 9 um')
print('V_T0 = {:g} V'.format(Vt0))
print('V_T1 = {:g} V'.format(Vt1))
print('gamma = {:g} V^1/2'.format(gamma))
print('kn = {:g} A/V^2'.format(kn))

#plot the data
figure(figsize=(8,6));
plot(vgs,Id_n_big*1e3,'b-',label='$I_D$, $V_{SB}=0$'); #pylab supports limited LaTeX syntax
plot(vgs,Id_n_big_vsb*1e3,'r--',label='$I_D$, $V_{SB}=2.835$');
xlabel('$V_{GS}$ (V)'); ylabel('$I_D$ (mA)');
legend(loc='best')

figure(figsize(8,6));
plot(vgs,sqrt(Id_n_big),'b-',label='$\sqrt{I_D}$, $V_{SB}=0$');
plot([Vt0,5],fit([Vt0,5]),'k:');
plot(vgs,sqrt(Id_n_big_vsb),'r--',label='$\sqrt{I_D}$, $V_{SB}=2.835$');
plot([Vt1,5],fit_v([Vt1,5]),'k:');
ylim(ymin=0)
xlabel('$V_{GS}$ (V)'); ylabel('$\sqrt{I_D}$ ($\sqrt{A}$)');
legend(loc='best')
show()
```

**3.** Draw a schematic of a CMOS inverter. The power supply of the circuit is $5\,\text{V}$, i.e. the source of the PMOS transistor is connected to $5\,\text{V}$. Choose minimum size for both length and width of PMOS and NMOS transistors. Save the circuit and print it out. Perform a DC analysis on the circuit by sweeping the input to the inverter from $0\,\text{V}$ to $5\,\text{V}$. Plot the output of the inverter. Label the threshold of the inverter on the plot.

**Solution:** The inverter circuit, simulation circuit, and DC simulation result are shown below.

```python
Image(filename='inv_ckt.png')
Image(filename='inv_sim_ckt.png')
vin,vout=loadtxt('inv_results.csv',delimiter=',',skiprows=1,unpack=True)
inv_func = lambda v: interp(v,vin,vout) - v #will be 0 when vout = vin
from scipy.optimize import newton
vth = newton(inv_func,0) #find zero crossing, starting the solver at 0
figure(figsize=(8,6));
plot(vin,vout,'b-');
plot(vin,vin,'k:');
plot([0,vth,vth],[vth,vth,0],'r--');
annotate(xy=(vth,vth),s='$V_{{th}}={:g}$'.format(vth));
xlabel('$V_{in}$'); ylabel('$V_{out}$');
show()
```
